```python
import pandas as pd
import random

# Club info with frequencies and participation range
club_events = {
    "Music Club": {"freq": 1, "min": 70, "max": 100},
    "BDACC": {"freq": 1, "min": 10, "max": 18},
    "FinWiz": {"freq": 1, "min": 90, "max": 120},
    "UNIQ": {"freq": 2, "min": 50, "max": 75},
    "LDC": {"freq": 3, "min": 45, "max": 65},
    "180DC": {"freq": 2, "min": 25, "max": 40},
    "BClub": {"freq": 2, "min": 20, "max": 35}
}

days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
schedule = []

used_slots = set()  # to avoid duplicate day-club entries

for club, info in club_events.items():
    chosen_days = random.sample(days, info["freq"])
    for day in chosen_days:
        while (day, club) in used_slots:
            day = random.choice(days)
        used_slots.add((day, club))
        participants = random.randint(info["min"], info["max"])
        schedule.append({
            "Day": day,
            "Club": club,
            "Event Type": "Club Activity",
            "Participants": participants
        })

# Create DataFrame and sort by Day
df_schedule = pd.DataFrame(schedule)
df_schedule = df_schedule.sort_values(by="Day", key=lambda x: pd.Categorical(x, categories=days, ordered=True)).reset_index(drop=True)

df_schedule
```
